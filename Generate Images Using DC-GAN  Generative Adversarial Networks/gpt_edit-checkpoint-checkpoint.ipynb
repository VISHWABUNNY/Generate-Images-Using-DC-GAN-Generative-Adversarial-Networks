```python
## load lib's
from keras.datasets import mnist
from keras.layers import *
from keras.layers import LeakyReLU
from keras.models import Sequential, Model
from keras.optimizers import Adam
import numpy as np 
import keras
import keras.losses as losses
import matplotlib.pyplot as plt
```

```python
(x_train,_),(_,_)= mnist.load_data()
```

```python
x_train.shape
```

```python
x_train=(x_train-127.5)/127.5
print(x_train.min())
print(x_train.max())
```

```python
TOTAL_EPOCHS=50
BATCH_SIZE=250
HALF_BATCH=128
NO_OF_BATCHES=int(x_train.shape[0]/BATCH_SIZE)
NOISE_DIM=100
adam=Adam(lr=2e-4,beta_1=0.5)
```

```python
# Generator Model: Upsampling
generator = Sequential()
generator.add(Dense(units=7*7*128, input_shape=(NOISE_DIM,)))
generator.add(Reshape((7, 7, 128)))
generator.add(LeakyReLU(0.2))
generator.add(BatchNormalization())

generator.add(Conv2DTranspose(64, (3, 3), strides=(2, 2), padding='same'))
generator.add(BatchNormalization())
generator.add(LeakyReLU(0.2))

generator.add(Conv2DTranspose(1, kernel_size=(3, 3), strides=(2, 2), padding='same', activation='tanh'))
generator.summary()
generator.compile(loss=losses.binary_crossentropy, optimizer=adam)
```

```python
# Discriminator Model: Downsampling
discriminator = Sequential()
discriminator.add(Conv2D(64, kernel_size=(3, 3), strides=(2, 2), padding='same', input_shape=(28, 28, 1)))
discriminator.add(LeakyReLU(alpha=0.2))

discriminator.add(Conv2D(128, kernel_size=(3, 3), strides=(2, 2), padding='same'))
discriminator.add(LeakyReLU(alpha=0.2))

discriminator.add(Flatten())
discriminator.add(Dense(100))
discriminator.add(LeakyReLU(0.2))

discriminator.add(Dense(1, activation='sigmoid'))
discriminator.compile(loss=losses.binary_crossentropy, optimizer=adam)
discriminator.summary()
```

```python
# Combined model
discriminator.trainable = False
gan_input = Input(shape=(NOISE_DIM,))
generator_img = generator(gan_input)
gan_output = discriminator(generator_img)
model = Model(gan_input, gan_output)
model.compile(loss=losses.binary_crossentropy, optimizer=adam)
```

```python
# Reshape x_train
x_train = x_train.reshape(-1, 28, 28, 1)
```

```python
# Function to display generated images
def display_images(samples=25):
    noise = np.random.normal(0, 1, size=(samples, NOISE_DIM))
    generated_img = generator.predict(noise)
    plt.figure(figsize=(10, 10))   
    for i in range(samples):
        plt.subplot(5, 5, i+1)
        plt.imshow(generated_img[i].reshape(28, 28), cmap="binary")     
        plt.axis('off')
    plt.show()
```

```python
# Training loop
discriminator_losses = []
generator_losses = []
for epoch in range(TOTAL_EPOCHS):
    epoch_discriminator_loss = 0.0
    epoch_generator_loss = 0.0

    for step in range(NO_OF_BATCHES):
        # Step 1: Train Discriminator
        discriminator.trainable = True
        idx = np.random.randint(0, 60000, HALF_BATCH)
        real_imgs = x_train[idx]
        noise = np.random.normal(0, 1, size=(HALF_BATCH, NOISE_DIM))
        fake_imgs = generator.predict(noise)
        real_y = np.ones((HALF_BATCH, 1)) * 0.9
        fake_y = np.zeros((HALF_BATCH, 1))

        discriminator_loss_real = discriminator.train_on_batch(real_imgs, real_y)
        discriminator_loss_fake = discriminator.train_on_batch(fake_imgs, fake_y)
        discriminator_loss = 0.5 * discriminator_loss_real + 0.5 * discriminator_loss_fake
        epoch_discriminator_loss += discriminator_loss

        # Step 2: Train Generator
        discriminator.trainable = False
        noise = np.random.normal(0, 1, size=(BATCH_SIZE, NOISE_DIM))
        ground_truth_y = np.ones((BATCH_SIZE, 1))

        generator_loss = model.train_on_batch(noise, ground_truth_y)
        epoch_generator_loss += generator_loss
        #----------------------------------------------------------------------------------------------------------

    print("Epoch {}, Disc loss {}, Generator Loss {}".format(epoch+1, epoch_discriminator_loss / NO_OF_BATCHES, epoch_generator_loss / BATCH_SIZE))

    discriminator_losses.append(epoch_discriminator_loss / NO_OF_BATCHES)
    generator_losses.append(epoch_generator_loss / NO_OF_BATCHES)

    if (epoch + 1) % 10 == 0:
        generator.save("generator.h5")
        display_images()
```
